```python
df_key = pd.read_csv('keywords.csv')
df_key.head()
```

```python
geo_data = {
  'Центр': ['москва', 'тула', 'ярославль'],
  'Северо-Запад': ['петербург', 'псков', 'мурманск'],
  'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
  }
```

```python
def geo_class(row):

  row['region'] = 'undefined'#изначально строка region со значением 'undefined'
  
  for key,value in geo_data.items():#цикл разбор словаря на ключи и значения
    for city in value:#цикл список на значения
      if city in row['keyword']:#проверка на вхождение частей списка в колонку кейворд
        row['region'] = key#когда вхождение тру, то происходит перезапись значения 'region' на ключ словаря
      else:
        pass#если вхождения нет, то ничего не происходит
  
  return row#функция должна вернуть всю строку целиком

  
```

```python
df_updated = df_key.apply(geo_class, axis = 1)#функция применяется ко всей строке датафрейма целиком, значение параметра ось = 1. Что-то в районе 1 мин код обрабатывал 100К строк
```

```python
year = list(range(1950,2011))
```

```python
df_ratings = pd.read_csv('ratings.csv')
df_ratings.info()
```

```python
df_movies = pd.read_csv('movies.csv')
df_movies.info()
```

```python
def production_year(row):

  row['prod_year'] = 1900

  for y in year:
    if str(y) in row['title']:
       row['prod_year'] = y
    else:
      pass
  return row
```

```python
df_year = df_movies.apply(production_year, axis = 1)
df_year.info()
```

```python
df_merged = df_ratings.merge(df_year, left_on='movieId', right_on='movieId')
df_merged.head()
```

```python
df_merged.groupby('prod_year').agg({'rating':'mean'}).sort_values('rating',ascending=False)
```
