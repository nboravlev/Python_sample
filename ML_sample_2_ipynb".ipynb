```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv ('spendings.csv', sep = ",")
df.head ()
```

Собраны данные за 4 месяца. Объектами являются позиции по вертикали: шоколад, мороженое, вино и т.д. Они имеют ряд признаков: 
RD_July = rainy days july = среднее количество дождливых дней в июле и по аналогии в остальные месяцы 
AT_July = average temperature = средняя температура в данном месяц;
Spendings = траты.

Цель: построить прогноз по тратам на Октябрь с помощью метода Регрессии и сравнить с фактическими данными.

```python
## формируем признаки и целевую переменную

X = df [['RD_July', 'AT_July', 'Spend_July', 'RD_Aug', 'AT_Aug',
       'Spend_Aug', 'RD_Sept', 'AT_Sept', 'Spend_Sept', 'RD_Oct', 'AT_Oct',]]

y = df ['Spend_Oct']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size = 0.3) ##раздел данных на выборки для проверки и обучения
```

```python
X_train
```

```python
X_test
```

```python
## создание и обучение модели

lr = LinearRegression ()
```

```python
lr.fit (X_train, y_train)
```

```python
lr.coef_
```

```python
## prediction

lr.predict (X_test)
```

```python
## сравнить значения "на глаз"

pd.DataFrame([y_test.values, lr.predict(X_test)]).T
```

```python
mean_squared_error(y_test, lr.predict(X_test))
```

```python
mean_squared_error(y_train, lr.predict(X_train))
```
